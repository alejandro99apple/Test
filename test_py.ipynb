```python
!pip install umap-learn
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap

# URL del conjunto de datos en UCI Machine Learning Repository
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

# Definir los nombres de las columnas
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Cargar los datos en un DataFrame
data = pd.read_csv(url, header=None, names=column_names)

# Separar características y etiquetas
X = data.iloc[:, :-1]  # Características
y = data.iloc[:, -1]   # Etiquetas

# Aplicación de PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X)

# Visualización con t-SNE
tsne = TSNE(n_components=2)
tsne_result = tsne.fit_transform(X)

# Aplicación de UMAP
umap_result = umap.UMAP(n_components=2).fit_transform(X)

# Visualización de resultados

# Visualización PCA
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=y.astype('category').cat.codes)
plt.title('PCA Result')
plt.xlabel('Component 1')
plt.ylabel('Component 2')

# Visualización t-SNE
plt.subplot(1, 3, 2)
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=y.astype('category').cat.codes)
plt.title('t-SNE Result')
plt.xlabel('Component 1')
plt.ylabel('Component 2')

# Visualización UMAP
plt.subplot(1, 3, 3)
plt.scatter(umap_result[:, 0], umap_result[:, 1], c=y.astype('category').cat.codes)
plt.title('UMAP Result')
plt.xlabel('Component 1')
plt.ylabel('Component 2')

plt.tight_layout()
plt.show()

```
